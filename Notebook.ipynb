```python

import requests
from IPython.display import Image, display

url = 'http://localhost:8000'

#  Request Overall Graphs
r1 = requests.get(f'{url}/overall')
overall = r1.json()['graphs']
# Request Lattitude Graphs
r2 = requests.get(f'{url}/overall/latitude')
latitude = r2.json()['graphs']
```

# UK Weather Summary
The data used for this project is from the Met Office [historic station data](https://www.metoffice.gov.uk/research/climate/maps-and-data/historic-station-data
) site, which provides 39,458 monthly readings from 37 stations across the UK.

The data includes:
  -  Max Tempurate (tmax) °C
  -  Min Tempurate (tmin) °C
  -  Air Frost (af) days
  -  Total Rainfall (rain) mm
  -  Total Sunshine (sun) hours

# Overall

## Tempurature
Overall annual tempurature has increased, with a +0.33°C per Century trend. This is to be expected as a result of climate change.

July and August are the hotest months.

```python
display(Image(filename=overall['avg_temp_trend']))
display(Image(filename=overall['total_temp']))
```

## Rainfall
Rainfall has a clear upward trend, +6.64mm per century, which is another expected result that follows from the tempurature increase: more water evaporates and comes back down as rain.

October, November, December and January have the highest rainfall. 

```python
display(Image(filename=overall['avg_rain_trend']))
display(Image(filename=overall['total_rainfall']))
```

## Sunshine
Sunshine is steady, with a relatively flat trend line.

The summer months, May, June, July and August have the most sunshine.

```python
display(Image(filename=overall['avg_sunshine_trend']))
display(Image(filename=overall['total_sunshine']))
```

# Latitude
A Stations latitude can have a notable affect on the weather.

## Tempurature
Unsuprisingly the further from the equator, the more the tempurature drops. The range from Lerwick, on the   Shetland Islands, to Camborne, in the south of Cornwall, gives us a range of ~4°C across the UK.

```python
display(Image(filename=latitude['temp']))
```

## Rainfall
Rainfall increases as you head north.

```python
display(Image(filename=latitude['rain']))
```

## Sunshine
The tilt of the earth leads to less sunshine at higher latitudes.

```python
display(Image(filename=latitude['sun']))
```
